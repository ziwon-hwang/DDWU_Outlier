```python
data_split = open('/content/drive/MyDrive/Colab Notebooks/NEWS_split_crawling_data.txt', 'r', encoding='utf-8').read()
data_split = [[data_split,1]]
print(data_split)
```

```python
from konlpy.tag import Okt

okt = Okt()

def tokenize(doc):
    return [t for t in okt.pos(doc, norm=True, stem=True)]      # normì€ ì •ê·œí™”, stemì€ ê·¼ì–´ë¡œ í‘œì‹œ

tokenize_docs = [(tokenize(row[0]), row[1]) for row in data_split]

with open('/content/drive/MyDrive/Colab Notebooks/tokenized_data', 'w', encoding="utf-8") as make_file:     # JSON íŒŒì¼ë¡œ ì €ì¥
    json.dump(tokenize_docs, make_file, ensure_ascii=False, indent="\t")
tokenize_docs[0]
```

# ì›Œë“œí´ë¼ìš°ë“œ

```python
# ë¹ˆë„ìˆ˜ ì„¸ê¸°

from collections import Counter
counter = Counter(tokenize_docs[0][0])
count_words = sorted(counter.items(), key=lambda x:-x[1])
count_words
```

```python
# ëª…ì‚¬ë§Œ ì¶”ì¶œ

noun_count = {x[0][0]:x[1] for x in count_words if x[0][1]=='Noun' and len(x[0][0])>=2 and 'ë™ë•' not in x[0][0]}
noun_count
```
